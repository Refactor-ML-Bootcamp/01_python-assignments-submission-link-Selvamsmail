# Visualization

```python
data.hist(bins=10,figsize=(10,10),sharey=True) # hist plot to find the distribution of the data.
```

```python
data.iloc[:,:-1].plot(kind='box',figsize=(10,6), title='sales') # ideas on outliers
```

```python
# feature to taget comparition
g = sns.pairplot(y_vars="Sales",x_vars=["Radio","Newspaper","TV"],data=data)
g.fig.set_size_inches(10,6)
```

```python
plt.figure(figsize=(10,6)) # line plot to see the unit of sales(trend...)
l = list(data.columns)
l = l[:-1]
for i in l:
  sns.lineplot(data=data, x="Sales",y=i)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),annot=True,cmap="Blues") # we see strong corelation with TV and SAlES
```
